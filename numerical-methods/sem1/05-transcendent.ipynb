```python
#5 (1/11)  найти корни уравнений с точностью 0.01 , выполнив не менее 5 итераций
#      Корни ищутся на отрезках длиной 1 и с целочисленными границами
#      Односторонние методы начинают попытку с левой границы           
#      При неудачной первой попытке - вычисления продолжить, начав с правой границы
#        ( -1 ;  0 ) метод половинного деления        2*x*x*x  + 3*x  + 2.2  = 0
#        (  1 ;  2 ) метод секущих (хорд)               x*x*x  + 3*x  - 8  = 0
#        (  1 ;  2 ) метод касательных (Ньютона)     0.5*x*x*x  + 11*x  - 18.75  = 0
#        ( -2 ; -1 ) метод  итераций                 4*x*x*x  + 7*x  + 44  = 0
```

```python
#Метод половинного деления
epsilon=0.01
from math import fabs
def F(x):
    return(float(2*x*x*x+3*x+2.2))
ab=[-1.0,0.0]
i=0
while  fabs(ab[0]-ab[1])>epsilon:
    c=(ab[0]+ab[1])/2
    i+=1
    print("{i}:a={a:.8f}  c={c:.8f}  b={b:.8f} f(a)={fa:.8f}  f(c)={fc:.8f}  f(b)={fb:.8f} "
          .format(i=i,a=ab[0],b=ab[1],c=c,fa=F(ab[0]),fb=F(ab[1]),fc=F(c)))
    if (F(ab[0])*F(c))<0:
        ab[1]=c
    elif F(ab[1])*F(c)<0:
        ab[0]=c
```

```python
#Метод секущих (хорд)
epsilon=0.01
from math import fabs
def F(x):
    return(float(x*x*x+3*x-8))
ab=[1.0,2.0]
i=0
while  fabs(ab[0]-ab[1])>epsilon:
    c=(ab[0]*F(ab[0])-ab[1]*F(ab[1]))/(F(ab[0])-F(ab[1]))
    i+=1
    print("{i}:a={a:.8f}  c={c:.8f}  b={b:.8f} f(a)={fa:.8f}  f(c)={fc:.8f}  f(b)={fb:.8f} "
          .format(i=i,a=ab[0],b=ab[1],c=c,fa=F(ab[0]),fb=F(ab[1]),fc=F(c)))
    if (F(ab[0])*F(c))<0:
        ab[1]=c
    elif F(ab[1])*F(c)<0:
        ab[0]=c
```

```python
#Метод касательных (Ньютона)
epsilon=0.01
def F(x):
    return(float(0.5*x*x*x+11.0*x-18.75))
def dF(x):
    return(float(1.5*x*x+11.0))
def ddF(x):
    return(float(3.0*x))
ab=[1.0,2.0]
if (F(ab[0])*ddF(ab[0])>0):
    a=ab[0]
else:
    a=ab[1]
    

for i in range(6):
    print("{i}:a={a:.8f}   f(a)={fa:.8f}  f'(a)={fc:.8f}  f''(a)={fb:.8f} "
          .format(i=i,a=a,fa=F(a),fb=ddF(a),fc=dF(a)))
    a=a-F(a)/dF(a)
```
