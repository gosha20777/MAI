```python
import numpy as np
from math import fabs
from math import atan
from math import cos
from math import sin

def eigenvalue(A, v):
    Av = A.dot(v)
    return v.dot(Av)

def power_iteration(A):
    A = A.copy()
    n, d = A.shape

    v = np.ones(d) / np.sqrt(d)
    ev = eigenvalue(A, v)
    iteration = 0

    while True:
        Av = A.dot(v)
        v_new = Av / np.linalg.norm(Av)

        ev_new = eigenvalue(A, v_new)
        if np.abs(ev - ev_new) < 0.0000000001:
            break

        v = v_new
        ev = ev_new
        iteration += 1

    return ev_new, v_new, iteration

def jokobi_3x3(A):
    a=A.copy()
    H_list=[]
    V=np.eye(3)
    for i in range(6):
        print ("Матрица A(",i,')')
        print (a)

        #поиск наибольшего числа
        k,m=0,1
        if fabs(a[0][2])>fabs(a[k][m]):
            k,m=0,2
        if fabs(a[1][2])>fabs(a[k][m]):
            k,m=1,2
        phi=1/2*atan(2*a[k][m]/(a[k][k]-a[m][m]))
        print(i,":",phi)
        H=np.eye(3)
        H[k][k]=cos(phi)
        H[m][m]=cos(phi)        
        H[k][m]=-sin(phi)    
        H[m][k]=sin(phi)
        H_list.append(H)
        a=np.dot(np.dot(H.T,a),H)
        
    for i in range (len(H_list)):
        print ("Матрица V(",i,')')
        print (V)
        V=np.dot(V,H_list[i])
    
    print ("Матрица A")
    print (a)
    print ("Матирица V")
    print (V)
```
