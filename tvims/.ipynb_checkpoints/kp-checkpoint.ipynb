### –†–µ—à–µ–Ω–∏–µ

1. –û–ø—Ä–µ–¥–µ–ª–∏–º –≤–∞—Ä–∏–∞—Ç–∏–≤–Ω—ã–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –∏ –∏–º–ø–æ—Ä—Ç–∏—Ä—É–µ–º –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã–µ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏:

- `n` - –∫–æ–ª–ª–∏—á–∫—Å—Ç–≤–æ –∏–∑–º–µ—Ä–µ–Ω–∏–π
- `h` - –≤—Ä–µ–º–µ–Ω–Ω–æ–π —à–∞–≥
- `v` - –ª–∏–Ω–µ–π–Ω–∞—è —Å–∫–æ—Ä–æ—Å—Ç—å
- `sigma` - —Å—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ
- `disp` - –¥–∏—Å–ø–µ—Ä—Å–∏—è
- `y0` - –Ω–∞—á–∞–ª—å–Ω–æ–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –æ—Ç –Ω–∞–±–ª—é–¥–∞—Ç–µ–ª—è (—Å–∞–º–æ–ª–µ—Ç –Ω–∞–¥ –Ω–∞–±–ª—é–¥–∞—Ç–µ–ª–µ–º)
- `t` - –≤–µ–∫—Ç–æ—Ä –≤—Ä–µ–º–µ–Ω–∏: $t_i = h*i$
- `x` - –≤–µ–∫—Ç–æ—Ä –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç —Å–∞–º–æ–ª–µ—Ç–∞ –≤ –∫–∞–∂–¥—ã–π –º–æ–º–µ–Ω—Ç –≤—Ä–µ–º–µ–Ω–∏ t: $x_i = v*t_t$
- `r` - –≤–µ–∫—Ç–æ—Ä –¥–∞–ª—å–Ω–æ—Å—Ç—å–∏ –∞ –∫–∞–∂–¥—ã–π –º–æ–º–µ–Ω—Ç –≤—Ä–µ–º–µ–Ω–∏
- `norm_errors` - –æ—à–∏–±–∫–∏ –Ω–∞–±–ª—é–¥–µ–Ω–∏—è: –∏—Ö –º–æ–∂–Ω–æ —Å–≥–µ–Ω–µ—Ä–∏—Ä–æ–≤–∞—Ç—å —Å –ø–æ–º–æ—à—å—é `np.random.normal(0, sigma)`
- `alpha` - —É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏

```python
# –∏–º–ø–æ—Ä—Ç –±–∏–±–ª–∏–æ—Ç–µ–∫
import numpy as np
import matplotlib.pyplot as plt
```

```python
# –≤–∞—Ä–∏–∞—Ç–∏–≤–Ω—ã–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã
n = 20
h = 2
v = 100
sigma = 100
disp = sigma**2
y0 = 10000
t = np.arange(0, h*n, h)
x = v*t
r = np.sqrt(x**2 + y0**2)
norm_errors = np.random.normal(0, sigma, t.shape[0])
alpha = 0.01
```

2. –ü–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫

```python
plt.figure(figsize=(10,10))
plt.plot(t, r, label = "—Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ r", color= "orange")
plt.scatter(t, r+norm_errors, label = "—Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ —Å –Ω–æ—Ä–º–∞–ª—å–Ω–æ –æ—à–∏–±–∫–∞–º–∏ r+norm_errors")
plt.legend()
plt.show()
```

3. –ø–æ—Å—Ç–æ—Ç—Ä–∏–º –Ω–∞ –≤–µ–∫—Ç–æ—Ä —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è –∏ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è —Å –æ—à–∏–±–∫–∞–º–∏

```python
print('—Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ:\n', r)
print('—Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ c –æ—à–∏–±–∫–∞–º–∏:\n', r+norm_errors)
```

### –†–µ—à–µ–Ω–∏–µ

1. –û–ø—Ä–µ–¥–µ–ª–∏–º —É—Ä–∞–≤–Ω–µ–Ω–∏–µ –ø—Ä–æ—Å—Ç–æ–π –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ (–≤ –º–∞—Ç—Ä–∏—á–Ω–æ–π —Ñ–æ—Ä–º–µ):

$$
Y = A\cdot\theta+W
$$

–≥–¥–µ:
- `ùëå` - –≤–µ–∫—Ç–æ—Ä –Ω–∞–±–ª—é–¥–µ–Ω–∏–π (n x 1)
- `ùê¥` - —Ä–µ–≥—Ä–µ—Å—Å–∏–æ–Ω–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞ (n x 2): –ø–µ—Ä–≤—ã–π —Å—Ç–æ–ª–±–µ—Ü - –µ–¥–∏–Ω–∏—á–Ω—ã–π, –≤—Ç–æ—Ä–æ–π - –∑–Ω–∞—á–µ–Ω–∏—è –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ ùë°
- `ùúÉ` - –≤–µ–∫—Ç–æ—Ä (2 x 1) –Ω–µ–∏–∑–≤–µ—Å—Ç–Ω—ã—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ (–≤–µ—Å–æ–≤ –º–æ–¥–µ–ª–∏)
- `ùëä` - –º–∞—Ç—Ä–∏—Ü–∞ (n x 1) –æ—à–∏–±–æ–∫ –Ω–∞–±–ª—é–¥–µ–Ω–∏–π
- `n` - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–∞–±–ª—é–¥–µ–Ω–∏–π

$$
A = \begin{bmatrix} 
    1 & t_{1}\\
    \vdots & \vdots\\
    1 & t_{n} 
    \end{bmatrix};\ \theta = \begin{bmatrix} 
    \theta_{1}\\
    \theta_{2} 
    \end{bmatrix} 
$$

2. –ó–∞–º–µ—Ç–∏–º —á—Ç–æ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ–µ  –∏  –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ–µ  —É—Å–ª–æ–≤–∏–µ–Ω–µ —Å–º–µ—â–µ–Ω–Ω–æ—Å—Ç–∏ –æ—Ü–µ–Ω–∫–∏ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è, —Ç.–∫. –Ω–µ—Ç —Å–∏—Å—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏—Ö –æ—à–∏–±–æ–∫: $$MW=0$$

3. –û—à–∏–±–∫–∏ –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã, –∑–Ω–∞—á–∏—Ç –º–æ–¥–µ–ª—å –æ–ø—Ç–∏–º–∞–ª—å–Ω–∞: 
$$
I_{n}= \begin{bmatrix} 
    1 & \dots & 1 \\
    \vdots & \ddots & \vdots \\
    0 & \dots  & 1 
    \end{bmatrix}\ \Rightarrow ùê∑ùëä_k=ùúé^2=ùëêùëúùëõùë†ùë°
$$

3. –û—Ü–µ–Ω–∏–º –≤–µ–∫—Ç–æ—Ä –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ (–≤–µ—Å–æ–≤) –ø–æ –º–µ—Ç–æ–¥—É –Ω–∞–∏–º–µ–Ω—å—à–∏—Ö –∫–≤–∞–¥—Ä–∞—Ç–æ–≤:
$$
\theta:(Y-A\cdot\theta)^2 = min_{\theta} 
$$

4. –ï—Å–ª–∏ $\exists(A^TA)^{-1}$:
$$
\hat{\theta} = \underset{\theta}{\mathrm{argmin}}(Y-A\theta)^2=(A^TA)^{-1}A^TY
$$

5. –ó–∞–¥–∞–¥–∏–º –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª:
$$
P(\hat{r}(t)-u_{\alpha}\sqrt{D_r(t)}\leq\hat{r}(t)+u_{\alpha}\sqrt{D_r(t)}=1-\alpha
$$

–ì–¥–µ:
- $u_{\alpha}$ - –∫–≤–∞–Ω—Ç–∏–ª—å –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
- $D_r(t)$ - –¥–∏—Å–ø–µ—Ä—Å–∏—è –ú–ù–ö-–æ—Ü–µ–Ω–∫–∏
- $\alpha$ - —É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏

6. –ó–∞–º–µ—Ç–∏–º —á—Ç–æ: $D_r(t) = M[(r-M[r])^2]\Rightarrow\hat{D}_r(t)=\frac{(Y-\hat{Y})^2}{n-2}$

–ì–¥–µ:
- $\hat{Y} = A\cdot\hat{\theta}$ - –æ—Ü–µ–Ω–∫–∞  –≤–µ–∫—Ç–æ—Ä–∞ –Ω–∞–±–ª—é–¥–µ–Ω–∏–π ùëå.

7. –í—ã–ø–æ–ª–Ω–∏–º –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã–µ –¥–ª—è —Ä–µ—à–µ–Ω–∏—è –∑–∞–¥–∞—á–∏ —Ä–∞—Å—Å—á–µ—Ç—ã, –ø—Ä–∏—á–µ–º:
- `n` - –∫–æ–ª–ª–∏—á–µ—Å—Ç–≤–æ –∏–∑–º–µ—Ä–µ–Ω–∏–π
- `A` - —Ä–µ–≥—Ä–µ—Å—Å–∏–æ–Ω–Ω–∞—è –º–∞—Ç—Ä–∏—Üa $A$
- `Y` - –≤–µ–∫—Ç–æ—Ä –Ω–∞–±–ª—é–¥–µ–Ω–∏–π $Y$
- `theta_hat` - –ú–ù–ö –æ—Ü–µ–Ω–∫–∞ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ $\hat{\theta}$
- `theta_1_hat` - –ü–µ—Ä–≤—ã–π –ø–∞—Ä–∞–º–µ—Ç—Ä $\hat{\theta}_1$
- `theta_2_hat` - –í—Ç–æ—Ä–æ–π –ø–∞—Ä–∞–º–µ—Ç—Ä $\hat{\theta}_2$
- `r_hat` - –ú–ù–ö-–æ—Ü–µ–Ω–∫–∏ $\hat{r}$
- `s` - –¥–∏—Å–ø–µ—Ä—Å–∏—è –ú–ù–ö-–æ—Ü–µ–Ω–∫–∏ $D_r(t)$
- `upper_bound`, `lower_bound` - –ì—Ä–∞–Ω–∏—Ü—ã –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞
- `E` - –æ—Å—Ç–∞—Ç–∫–∏ (–æ—à–∏–±–∫–∞) –ú–ù–ö-–æ—Ü–µ–Ω–∫–∏
- `u_a` - –∫–≤–∞–Ω—Ç–∏–ª—å –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è

```python
n = r.size
u_a = 2.326
A = np.matrix(np.concatenate((np.ones(shape = (n, 1), dtype = int), t.reshape(n,1)), axis=1))
Y = np.matrix(r+norm_errors).T
theta_hat = np.dot(np.dot(np.linalg.inv(np.dot(A.T,A)),A.T), Y)
theta_1_hat = float(theta_hat[0,0])
theta_2_hat = float(theta_hat[1,0])
r_hat = theta_1_hat + t * theta_2_hat
s = np.power(np.sum(r - r_hat),2)/(n-2)
upper_bound = r_hat + u_a*np.sqrt(s)
lower_bound = r_hat - u_a*np.sqrt(s)
E = r+norm_errors - r_hat
```

8. –ü–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫

```python
plt.figure(figsize=(14,14))
plt.title("–ú–æ–¥–µ–ª—å –ø—Ä–æ—Å—Ç–æ–π –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏")
plt.scatter(t,r, label = "–∏—Å—Ç–∏–Ω–Ω–∞—è –¥–∞–ª—å–Ω–æ—Å—Ç—å", color = 'green')
plt.scatter(t,r+norm_errors, label = "–Ω–∞–±–ª—é–¥–µ–Ω–∏—è —Å –æ—à–∏–±–∫–∞–º–∏", color = 'red')
plt.plot(t, r_hat, label = "–ú–ù–ö-–æ—Ü–µ–Ω–∫–∞", color = 'blue')
plt.plot(t, upper_bound, label = "–¥–æ–≤—Ä–∏—Ç–µ–ª—å–Ω—ã–µ –≥—Ä–∞–Ω–∏—Ü—ã", color = 'purple')
plt.plot(t, lower_bound, color = 'purple')
plt.legend()
plt.show()
```

### –†–µ—à–µ–Ω–∏–µ

1. –†–µ—à–µ–Ω–∏–µ –∞–Ω–∞–ª–æ—à–∏—á–Ω–æ –ø—Ä–µ–¥—ã–¥—É—â–µ–º—É –∑–∞–¥–∞–Ω–∏—é —Å —Ç–æ–π –ª–∏—à–± —Ä–∞–∑–Ω–∏—Ü–µ–π —á—Ç–æ –º–∞—Ç—Ä–∏—Ü–∞ $A$ –∏ $\theta$ —Ç–µ–ø–µ—Ä—å –∏–º–µ–µ—é—Ç —Ä–∞–∑–º–µ—Ä (n x 3) –∏ (3 x 1) —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ:

$$
A = \begin{bmatrix} 
    1 & t_{1} & \frac{t_{1}^2}{2}\\
    \vdots & \vdots & \vdots\\
    1 & t_{n} & \frac{t_{n}^2}{2}
    \end{bmatrix};\ \theta = \begin{bmatrix} 
    \theta_{1}\\
    \theta_{2}\\
    \theta_{3}
    \end{bmatrix} 
$$

2. –í—ã–ø–æ–ª–Ω–∏–º –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã–µ –¥–ª—è —Ä–µ—à–µ–Ω–∏—è –∑–∞–¥–∞—á–∏ —Ä–∞—Å—Å—á–µ—Ç—ã, –ø—Ä–∏—á–µ–º:
- `n` - –∫–æ–ª–ª–∏—á–µ—Å—Ç–≤–æ –∏–∑–º–µ—Ä–µ–Ω–∏–π
- `A` - —Ä–µ–≥—Ä–µ—Å—Å–∏–æ–Ω–Ω–∞—è –º–∞—Ç—Ä–∏—Üa $A$
- `Y` - –≤–µ–∫—Ç–æ—Ä –Ω–∞–±–ª—é–¥–µ–Ω–∏–π $Y$
- `theta_hat` - –ú–ù–ö –æ—Ü–µ–Ω–∫–∞ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ $\hat{\theta}$
- `theta_1_hat` - –ü–µ—Ä–≤—ã–π –ø–∞—Ä–∞–º–µ—Ç—Ä $\hat{\theta}_1$
- `theta_2_hat` - –í—Ç–æ—Ä–æ–π –ø–∞—Ä–∞–º–µ—Ç—Ä $\hat{\theta}_2$
- `theta_3_hat` - –¢—Ä–µ—Ç–∏–π –ø–∞—Ä–∞–º–µ—Ç—Ä $\hat{\theta}_3$
- `r_hat` - –ú–ù–ö-–æ—Ü–µ–Ω–∫–∏ $\hat{r}$
- `s` - –¥–∏—Å–ø–µ—Ä—Å–∏—è –ú–ù–ö-–æ—Ü–µ–Ω–∫–∏ $D_r(t)$
- `upper_bound`, `lower_bound` - –ì—Ä–∞–Ω–∏—Ü—ã –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞
- `E2` - –æ—Å—Ç–∞—Ç–∫–∏ (–æ—à–∏–±–∫–∞) –ú–ù–ö-–æ—Ü–µ–Ω–∫–∏
- `u_a` - –∫–≤–∞–Ω—Ç–∏–ª—å –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏

```python
n = r.size
A = np.matrix(np.concatenate((np.concatenate((np.ones(shape = (n, 1), dtype = int), t.reshape(n,1)), axis=1), ((np.power(t,2))/2).reshape(n,1)), axis = 1))
Y = np.matrix(r+norm_errors).T
theta_hat = np.dot(np.dot(np.linalg.inv(np.dot(A.T,A)),A.T), Y)
theta_1_hat = float(theta_hat[0,0])
theta_2_hat = float(theta_hat[1,0])
theta_3_hat = float(theta_hat[2,0])
r_hat = theta_1_hat + t * theta_2_hat + np.power(t,2) * theta_3_hat/2
s = np.power(np.sum(r-r_hat),2)/(n-2)
upper_bound = r_hat + u_a*np.sqrt(s)
lower_bound = r_hat - u_a*np.sqrt(s)
E2 = r+norm_errors - r_hat
```

3. –ü–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫

```python
plt.figure(figsize=(14,14))
plt.title("–ú–æ–¥–µ–ª—å –ø–∞—Ä–∞–±–æ–ª–∏—á–µ—Å–∫–æ–π –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏")
plt.scatter(t,r, label = "–∏—Å—Ç–∏–Ω–Ω–∞—è –¥–∞–ª—å–Ω–æ—Å—Ç—å", color = 'green')
plt.scatter(t,r+norm_errors, label = "–Ω–∞–±–ª—é–¥–µ–Ω–∏—è —Å –æ—à–∏–±–∫–∞–º–∏", color = 'red')
plt.plot(t, r_hat, label = "–ú–ù–ö-–æ—Ü–µ–Ω–∫–∞", color = 'blue')
plt.plot(t, upper_bound, label = "–¥–æ–≤—Ä–∏—Ç–µ–ª—å–Ω—ã–µ –≥—Ä–∞–Ω–∏—Ü—ã", color = 'purple')
plt.plot(t, lower_bound, color = 'purple')
plt.legend()
plt.show()
```

### –†–µ—à–µ–Ω–∏–µ

1. –í–≤–µ–¥–µ–º –Ω—É–ª–µ–≤—É—é –≥–µ–ø–æ—Ç–∏–∑—É $H_0$: $\theta_{3}=0$
2. –¢–æ–≥–¥–∞ –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–∏–∑–∞ $H_1$: $\theta_{3}\neq0$
3. –í–≤–µ–¥–µ–º —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É: $T = \frac{\hat{\theta}_3}{\sqrt{D[\hat{\theta}_3]}}$
4. –ö—Ä–∏—Ç–∏—á–µ—Å–∫–∞—è –æ–±–ª–∞—Å—Ç—å: $P(-u_{1-\frac{\alpha}{2}}\leq T\leq u_{1-\frac{\alpha}{2}}) = 1 - \alpha$
5. –ò–∑ —Å–≤–æ–π—Å—Ç–≤ —Ç–µ–æ—Ä–µ–º—ã –ì–∞—É—Å—Å–∞-–ú–∞—Ä–∫–æ–≤–∞ —Å–ª–µ–¥—É–µ—Ç: $M[\hat{\theta}_3] = \theta_3$
6. –ü—Ä–æ–∏–∑–≤–µ–¥–µ–º —Ä—ã—á–∏—Å–ª–µ–Ω–∏—è –¥–ª—è –¥–∞–Ω–Ω–æ–π –∑–∞–¥–∞—á–∏:
$$
T\mid_{H_0} = \frac{\hat{\theta}_3}{\sqrt{(\hat{\theta}_3-0)^2}};\ \alpha = 0.01
$$

$$
P(-u_{0.995}\leq \frac{\hat{\theta}_3}{\sqrt{(\hat{\theta}_3-0)^2}} \leq u_{0.995}) = 0.99 \Rightarrow P(-2.58\leq 1 \leq 2.58)
$$

–ü—Ä–∏ –≤—ã–ø–æ–ª–Ω–µ–Ω–∏–∏ –≥–∏–ø–æ—Ç–µ–∑—ã $H_0$, —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ $T$ –≤—Ö–æ–¥–∏—Ç –≤ –∫—Ä–∏—Ç–∏—á–µ—Å–∫—É—é –æ–±–ª–∞—Å—Ç—å, —Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ –æ–Ω–∞ –≤–µ—Ä–Ω–∞
–Ω–∞ —É—Ä–æ–≤–Ω–µ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ $\alpha=0.01$. –ó–Ω–∞—á–∏—Ç $\theta_{3}=0$.

## –†–µ—à–µ–Ω–∏–µ

1. –ü–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –æ—Å—Ç–∞—Ç–∫–æ–≤ –º–æ–¥–µ–ª–∏ –ø—Ä–æ—Å—Ç–æ–π –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ —Å –ø–æ–º–æ—â—å—é `plt.scatter(t,E)`
2. –° –ø–æ–º–æ—â—å—é `scipy.stats.norm.pdf(x)` –ø–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ $N(0, \sigma^2)$
3. –° –ø–æ–º–æ—â—å—é plt.hist –ø–æ—Å—Ç—Ä–æ–∏–º –Ω–∞ —Ç–æ–º –∂–µ –≥—Ä–∞—Ñ–∏–∫–µ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É –ø–æ –≤–µ–∫—Ç–æ—Ä—É –æ—Å—Ç–∞—Ç–∫–æ–≤

```python
import scipy.stats as stats
plt.figure(figsize=(14,14))
plt.scatter(t,E)
plt.title("–ì—Ä–∞—Ñ–∏–∫ –æ—Å—Ç–∞—Ç–∫–æ–≤ –º–æ–¥–µ–ª–∏ –ø—Ä–æ—Å—Ç–æ–π –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏")
plt.show()

x = np.linspace(stats.norm.ppf(0.01),
                stats.norm.ppf(0.99), 100)

plt.figure(figsize=(14,14))
plt.plot(x, stats.norm.pdf(x))
plt.hist(E/100, density=True, histtype='stepfilled', alpha=0.5)
plt.title("–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –æ—Å—Ç–∞—Ç–∫–æ–≤ –∏ –≥—Ä–∞—Ñ–∏–∫ –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ $N(0, \sigma^2)$")
plt.show()
```

```python
plt.figure(figsize=(14,14))
plt.scatter(t, E2)
plt.title("–ì—Ä–∞—Ñ–∏–∫ –æ—Å—Ç–∞—Ç–∫–æ–≤ –º–æ–¥–µ–ª–∏ –ø–∞—Ä–∞–±–æ–ª–∏—á–µ—Å–∫–æ–π –ø—Ä–æ—Å—Ç–æ–π –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏")
plt.show()

# 
x = np.linspace(stats.norm.ppf(0.01),
                stats.norm.ppf(0.99), 100)

plt.figure(figsize=(14,14))
plt.plot(x, stats.norm.pdf(x))
plt.hist(E2/100, density=True, histtype='stepfilled', alpha=0.5)
plt.title("–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –æ—Å—Ç–∞—Ç–∫–æ–≤ –∏ –≥—Ä–∞—Ñ–∏–∫ –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ $N(0, \sigma^2)$")
plt.show()
```
